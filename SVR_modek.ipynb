### Implementing the SVM

```python
from sklearn import svm 

clf = svm.SVR(kernel = 'linear', C = 10)

clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

### Hyperparameter Optimization

Here we choose to optimize four hyperparameters for a support vector machine classifier, these are:
- **C:** regularization parameter
- **kernel:** pecifies the kernel type to be used in the algorithm (linear, poly, rbf, sigmoid, precomputed)
- **poly**: degree of the polynomial kernel function (poly)
- **gamma**: kernel coefficient for rbf, poly and sigmoid

#### Defining hyperparameter space

```python
hyperparameters = {"C":[0.001, 0.01, 0.1, 1, 10, 100] , 
                  "kernel" : ["linear", "poly", "rbf", "sigmoid"], 
                  "degree" : [1, 2, 3, 4],
                 "gamma": [0.0001, 0.001, 0.01, 0.1, 1, 10]}
```

#### Manual Search

```python
# set model hyperparameters
clf_rbf = svm.SVR(kernel = 'rbf', C = 10, gamma = 0.1)

# fit the model to the training data
clf_rbf.fit(X_train, y_train)

#generate a set of predictions
y_pred = clf_rbf.predict(X_test)

# model evaluation
clf_rbf.score(X_test,y_test)
```

#### Random search

```python
from sklearn.model_selection import RandomizedSearchCV

# perform random search
clf_rs = RandomizedSearchCV(svm.SVR(), hyperparameters, cv = 3, scoring = "f1", n_iter = 20)
random_search = clf_rs.fit(X_train, y_train)

# identify best parameters from random search
random_search.best_params_
```

```python
# set model hyperparameters
model_svr1 = svm.SVR(kernel='linear',gamma=0.0001,degree=1,C=0.1)
model_svr1.fit(X_train,y_train)
# model evaluation
model_svr1.score(X_test,y_test)
```

#### Grid search

```python
from sklearn.model_selection import GridSearchCV

# perform grid search
clf_gs = GridSearchCV(svm.SVR(), hyperparameters, cv = 3, scoring = "f1")
grid_search = clf_gs.fit(X_train, y_train)

# identify best parameters from grid search
grid_search.best_params_
```

```python
# set model hyperparameters
model_svr2 = svm.SVR(kernel='linear',gamma=0.0001,degree=1,C=0.1)
model_svr2.fit(X_train,y_train)
# model evaluation
model_svr2.score(X_test,y_test)
```

#### Bayesian Optimization

```python
from skopt import BayesSearchCV

# perform bayesian optimization
clf_bo = BayesSearchCV(svm.SVC(), hyperparameters, cv =3, scoring = "f1", n_iter = 20)
bayes_search = clf_bo.fit(X_train, y_train)

# identify best parameters from bayesian optimization
bayes_search.best_params_
```

```python
# set model hyperparameters
model_svr3 = svm.SVR(kernel='linear',gamma=0.0001,degree=1,C=0.1)
model_svr3.fit(X_train,y_train)
# model evaluation
model_svr3.score(X_test,y_test)
```
